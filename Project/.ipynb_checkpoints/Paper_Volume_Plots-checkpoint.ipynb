# ApproxNEB Data Analysis & Processing

```python
from pymatgen.analysis.local_env import VoronoiNN, CrystalNN
vnn = VoronoiNN()
cnn = CrystalNN()
```

```python
def process(wf_uuid,hop_key):
    doc = aneba_store.query_one({"wf_uuid":wf_uuid})
    structs = [Structure.from_dict(s) for s in doc["data"][hop_key]["output_structures"]]
    indexes = doc["data"][hop_key]["relax_site_indexes"]
    vols = [sum([v["volume"] for v in vnn.get_voronoi_polyhedra(struct,i).values()]) for i,struct in zip(indexes,structs)]
    
    cns = []
    for n,struct in zip(indexes,structs):
        # crystal nn
        cn = cnn.get_cn(struct,n)
        cns.append(cn)
    
    return {"wf_uuid":wf_uuid,
            "vv":vols,
            "cn":cns,
            "energies":doc["data"][hop_key]["energies"],
            "wi_sites":doc["data"][hop_key]["wi_sites"],
            "structs":doc["data"][hop_key]["output_structures"],
            "doc":doc
           }
```

```python
results = {"43005_Mg":process("4599dc51-efd2-4b32-ac51-0cc982f08bcd","2+3"),# NASICON, "43005_Mg"
           "119666_Mg":process("9da400ce-b10e-496b-89f4-404c7cda55bb","0+2"), # Birnessite, "119666_Mg"
           "128543_Mg":process("7bd62bf7-a09c-4bf0-a02e-1224e00663b2","3+6"), # Tavorite, "128543_Mg"
           "113091_Mg":process("938826bf-676d-473e-b98f-0c3c9f842eed","3+4") #MnO2 spinel, "113091_Mg" -> don't have mg struct stored
          }
r2 = process("7bd62bf7-a09c-4bf0-a02e-1224e00663b2","3+5") # Tavorite, "128543_Mg"
for k in ["vv","cn","energies","wi_sites","structs"]:
    results["128543_Mg"][k].extend(r2[k])
    
print(results.keys())
```

# Plotting

```python
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

```python
def get_midpoints(array):
    """
    helper function that returns a list of midpoints between each
    element of list, but beginning return list with 0, ending with original last
    """
    output = [0]
    last = array[-1]
    for i in range(len(array)-1):
        a = array[i]
        b = array[i+1]
        mid = (a+b)/2
        output.append(mid)
    output.append(last)
    return output
```

```python
def get_distances(site_list,num_hop_points=9):
    """
    site_list (list): list of PeriodicSite objects
    returns: np.array of distances
    """
    distances = np.array([])
    n0_site = site_list[0]
    for n, n1_site in enumerate(site_list):
        if n0_site is None or n1_site is None:
            step_distance = 0
        if n % num_hop_points == 0:
            step_distance = 0
        else:
            step_distance = n1_site.distance(n0_site)
        distances = np.append(distances, step_distance)
        n0_site = n1_site

    #example usage to get distances for list of structures in a hop
    #distances = get_distances(wi_sites)
    #path_distances = distances.cumsum()  # note distances is a np.array
    #path_distances_norm = 100 * path_distances / path_distances.max()

    return distances
```

```python
def plot(x,y1,y2,title="",x_label="Image Number",y1_label="$\Delta$E (meV)",y2_label="Site Volume (A$^3$)"):
    y1c = "k" # energy axis color, y1
    y2c = "b" # volume axis color, y2
    
    plt.rcParams.update({'font.size': 18})
    plt.rc('xtick',labelsize=15)
    plt.rc('ytick',labelsize=15)
    fig1, ax1 = plt.subplots()
    ax1.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
    ax2 = ax1.twinx()

    plt.title(title,fontsize=18)
    ax1.set_xlabel(x_label,fontsize=18)
    ax1.set_ylabel(y1_label, color=y1c,fontsize=18)
    ax2.set_ylabel(y2_label, color=y2c,fontsize=18)
    ax1.tick_params(axis='both', which='major', labelsize=18)
    ax2.tick_params(axis='both', which='major', labelsize=18, color=y2c, labelcolor=y2c)

    ax1.plot(x,y1, color=y1c, linestyle="-", marker="o", markersize=6)
    ax2.plot(x,y2, color=y2c, linestyle="--", marker="s")
    plt.xlim(0, x[-1])
    ax1.set_ylim(0,1020)
    ax2.set_ylim(9,16)

    # adding background colors corresponding to coord_num
    colors = [0, 1, "#b0ffe1", "#ffc4c4", "#ffe3c4", "#dcc4ff", "#dcffc4", "#c4f2ff", '#fffec7']
    x_imgnum = get_midpoints(x)
    used_coord_nums = []
    for i in range(len(x_imgnum)-1):
        if plot_cns[i] in used_coord_nums:
            l = ""
        else:
            l = plot_cns[i]
            used_coord_nums.append(plot_cns[i])
        if plot_cns[i] is not None:
            v = ax1.axvspan(x_imgnum[i], x_imgnum[i+1], facecolor=colors[int(plot_cns[i])], 
                            alpha=0.9, label=l)
    ax1.legend(loc="upper right", bbox_to_anchor=(1.45, 0.7)) # NAISCON
    #ax1.legend(loc="upper right", bbox_to_anchor=(1.45, 0.43)) # birnessite
    #ax1.legend(loc="upper right", bbox_to_anchor=(1.45, 0.55)) # tavorite
    #ax1.legend(loc="upper right", bbox_to_anchor=(1.45, 0.31)) #spinel
```

```python
data = results["119666_Mg"]

energies = np.array(data["energies"],dtype=float)
plot_energies = (energies - min(energies))*1000
energy_barrier = max(plot_energies) - min(plot_energies)
plot_im_num = [i for i in range(len(energies))]
distances = get_distances([PeriodicSite.from_dict(d) for d in data["wi_sites"]])
path_distances = distances.cumsum()  # note distances is a np.array
path_distances_norm = 100 * path_distances / path_distances.max()

plot_vs = data["vv"]
plot_cns = data["cn"]
plot_cns.append(plot_cns[-1])

print("Barrier",round(energy_barrier),"meV")
print(path_distances[-1])

# image number
plot(plot_im_num,plot_energies,plot_vs)

# distances normalized
plot(path_distances,plot_energies,plot_vs,x_label="Path Distance (A)")

# distances normalized
plot(path_distances_norm,plot_energies,plot_vs,x_label="Path Distance (%)")
```
